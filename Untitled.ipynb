```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sphere_points(r, N):
    points = []
    for _ in range(N):
        z = np.random.uniform(-r, r)
        phi = np.random.uniform(0, 2 * np.pi)
        xy_radius = np.sqrt(r**2 - z**2)
        x = xy_radius * np.cos(phi)
        y = xy_radius * np.sin(phi)
        points.append((x, y, z))
    return np.array(points)

# Parameters
radius = 1.0
num_points = 1000

# Generate points
points = generate_sphere_points(radius, num_points)

# Optional: plot the points
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=1)
ax.set_box_aspect([1,1,1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def fibonacci_sphere_points(N, radius=1.0):
    points = []
    golden_angle = np.pi * (3 - np.sqrt(5))
    for i in range(N):
        z = 1 - (2 * i) / (N - 1)
        theta = golden_angle * i
        xy_radius = np.sqrt(1 - z**2)
        x = np.cos(theta) * xy_radius
        y = np.sin(theta) * xy_radius
        points.append((radius * x, radius * y, radius * z))
    return np.array(points)

# Parameters
N = 500
r = 10.0
points = fibonacci_sphere_points(N, r)

# Save points
np.savetxt("points_coor-1.dat", points, fmt="%.6f", delimiter="\t", header="# x_coor y_coor z_coor", comments="")

# Plot 3D "ball"
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Optional: color by z for lighting effect
colors = points[:, 2]  # z-value as color
ax.scatter(points[:, 0], points[:, 1], points[:, 2],
           c=colors, cmap='viridis', s=4, edgecolors='none')

# Equal aspect ratio for a real sphere look
ax.set_box_aspect([1, 1, 1])
ax.axis('off')  # Hide axes for cleaner look

# Optional: improve perspective
ax.view_init(elev=30, azim=30)

# Save figure
plt.savefig("points-1.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load dot coordinates
points = np.loadtxt("points_coor-1.dat")
x, y, z = points[:, 0], points[:, 1], points[:, 2]

# Set up figure
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 1. Draw white sphere surface (wireframe or mesh)
u, v = np.linspace(0, 2 * np.pi, 100), np.linspace(0, np.pi, 100)
xs = np.outer(np.cos(u), np.sin(v))
ys = np.outer(np.sin(u), np.sin(v))
zs = np.outer(np.ones_like(u), np.cos(v))

ax.plot_surface(xs, ys, zs, color='white', alpha=1.0, edgecolor='lightgray', linewidth=0.3, zorder=1)

# 2. Overlay black dots
ax.scatter(x, y, z, color='black', s=6, zorder=2)

# Aesthetics
ax.set_box_aspect([1, 1, 1])
ax.axis('off')
ax.view_init(elev=30, azim=35)

# Save and show
plt.savefig("white_sphere_black_dots.png", dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
